```python
import pandas as pd
import numpy as np
from itertools import groupby
```

```python
SQR = pd.read_csv('data/ngram_allTime_phone_impGT0.csv', encoding = "ISO-8859-1")
SQR['Clicks'] = SQR['Clicks'].convert_objects(convert_numeric=True)
SQR['Impressions'] = SQR['Impressions'].convert_objects(convert_numeric=True)
SQR['ConversionValue'] = SQR['ConversionValue'].convert_objects(convert_numeric=True)
SQR['Conversions'] = SQR['Conversions'].replace('[\$,]', '', regex=True).astype(float)
SQR['Cost'] = SQR['Cost'].replace('[\$,]', '', regex=True).astype(float)
```

```python
SQR
```

```python
from sklearn.feature_extraction.text import CountVectorizer

word_vectorizer = CountVectorizer(ngram_range=(1,2), analyzer='word')

sparse_matrix = word_vectorizer.fit_transform(SQR['Search term'])
sparse_matrix = sparse_matrix.tocoo()
#frequencies = sum(sparse_matrix).toarray()[0]


```

```python
output = []
id2vocab = dict((v,k) for k,v in word_vectorizer.vocabulary_.items())
zip_rc = sorted(zip(sparse_matrix.col, sparse_matrix.row), key=lambda x: x[0]) # group by column (vocab)
count = np.ravel(sparse_matrix.sum(axis=0)) # simple sum column for count
for g in groupby(zip_rc, key=lambda x: x[0]):
    index = g[0]
    bigram = id2vocab[index]
    loc = [g_[1] for g_ in g[1]]
    SQRInd = SQR.iloc[loc]
    c = count[index]
    impressions = SQRInd['Impressions'].sum()
    clicks = SQRInd['Clicks'].sum()
    cost = SQRInd['Cost'].sum()
    conv_value = SQRInd['ConversionValue'].sum()
    output.append([index, bigram, c, loc, impressions, clicks, cost, conv_value])
```

```python
SQRTerms = pd.DataFrame(output, columns = ['index', 'SQTerms', 'frequecy', 'documents', 'impressions', 'clicks', 'cost', 'conv_value'])
SQRTerms.sort_values(['cost', 'SQTerms'], ascending=[0, 0]).head(n=1000)
#SQRTerms.sort_values(['cost', 'SQTerms'], ascending=[0, 0]).head(n=100000)
```
